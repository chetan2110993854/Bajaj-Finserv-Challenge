```python
import json

def process_medication_data(file_path):
    with open(file_path) as file:
        data = json.load(file)


    for item in data:
        appointment_id = item.get('appointmentId')
        patient_details = item.get('patientDetails')
        if patient_details:
            first_name = patient_details.get('firstName')
            last_name = patient_details.get('lastName')
            gender = patient_details.get('gender')
            birth_date = patient_details.get('birthDate')
            phone_number = item.get('phoneNumber')
            medicines = item.get('consultationData', {}).get('medicines', [])


            for med in medicines:
                medicine_name = med.get('medicineName')
                frequency = med.get('frequency')
                duration = med.get('duration')
                instruction = med.get('instruction')
                is_active = med.get('isActive')




                print(f"Patient: {first_name} {last_name} | Gender: {gender} | Birth Date: {birth_date}")
                print(f"Phone Number: {phone_number}")
                print(f"Medication - Name: {medicine_name} | Frequency: {frequency} | Duration: {duration} | Instruction: {instruction} | Active: {is_active}")
                print("--------------------------------------------------")

# File paths for the provided files
file_path_1 = "/content/drive/MyDrive/DataEngineeringQ2.json"


process_medication_data(file_path_1)

```

```python
import matplotlib.pyplot as plt

# Data from the search results
appointments_data = [
    {"gender": "M", "appointments": 3},
    {"gender": "F", "appointments": 2},
]

# Extracting gender and appointments data
genders = [data["gender"] for data in appointments_data]
appointments = [data["appointments"] for data in appointments_data]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(appointments, labels=genders, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adding a title
plt.title('Appointments Distribution by Gender')

# Display the pie chart
plt.show()
```
